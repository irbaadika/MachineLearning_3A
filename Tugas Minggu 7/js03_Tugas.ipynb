```python
# IMPORT LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
# Tampilkan tabel mushrooms

tabel = pd.read_csv('dataset/mushrooms.csv')
tabel.head()
```

```python
tabel_baru = LabelEncoder()

for column in tabel.columns:
    tabel[column] = tabel_baru.fit_transform(tabel[column])

tabel.head()
```

```python
x = tabel.iloc[:,1:]
y = tabel.iloc[:,0]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
entropy = DecisionTreeClassifier(criterion='entropy')
gini = DecisionTreeClassifier(criterion='gini')

entropy.fit(x_train, y_train)
enteropy_pred = entropy.predict(x_test)

gini.fit(x_train, y_train)
gini_pred = gini.predict(x_test)

enteropy_accuracy = accuracy_score(y_test, enteropy_pred)
gini_accuracy = accuracy_score(y_test, gini_pred)

print(f'Entropy Accuracy: {enteropy_accuracy}')
print(f'Gini Accuracy: {gini_accuracy}')
```

```python
log_reg = LogisticRegression(solver="liblinear")
log_reg.fit(x_train, y_train)
hasil_log_reg = log_reg.score(x_test, y_test)

print(f'Hasil Logistic Regresion: {hasil_log_reg}')
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
gasil_gnb = gnb.score(x_test, y_test)
print(f'Hasil Naive Bayes: {gasil_gnb}')
```
